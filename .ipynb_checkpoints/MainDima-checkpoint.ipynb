# Предобработка

```python
subs_features_train = pd.read_csv('train/subs_features_train.csv', sep=';', decimal=',',encoding = "windows-1251")
subs_csi_train = pd.read_csv('train/subs_csi_train.csv', sep=';', decimal=',',encoding = "windows-1251")

print(subs_features_train.shape)
print(subs_csi_train.shape)
```

```python
subs_features_train = subs_features_train.drop_duplicates('SK_ID')
print(subs_features_train.shape)
```

```python
train = subs_features_train
target_table_train = subs_csi_train
```

```python
subs_features_test = pd.read_csv('test/subs_features_test.csv', sep=';', decimal=',',encoding = "windows-1251")
subs_csi_test = pd.read_csv('test/subs_csi_test.csv', sep=';', decimal=',',encoding = "windows-1251")

print(subs_features_test.shape)
print(subs_csi_test.shape)
```

```python
subs_features_test = subs_features_test.drop_duplicates('SK_ID')
print(subs_features_test.shape)
```

```python
test = subs_features_test
target_table_test = subs_csi_test
```

```python
train.fillna(train.mean())
test.fillna(test.mean())
```

```python
# рандом фичи
features = [
#             'SK_ID',
            'ITC',
#             'VAS',
#             'COM_CAT#1',
            'COM_CAT#2',
#             'COST',
            'COM_CAT#21',
#             'COM_CAT#22',
#             'COM_CAT#23',
#             'COM_CAT#25',
#             'RENT_CHANNEL',
#             'COM_CAT#26', 
            'COM_CAT#27',        
#             'ACT',
           ]
```

```python
y = target_table_train.CSI
X = train[features] #[0:4222]
from sklearn.model_selection import train_test_split
X_train, X_eval, Y_train, Y_eval = train_test_split(X, y, test_size = 0.1, random_state=42)
```

# Оценка фич

```python
from sklearn.ensemble import ExtraTreesClassifier
x = -1
feat_model = ExtraTreesClassifier(random_state=42)
feat_model.fit(X_train, Y_train)
lst = feat_model.feature_importances_
lst
# sorted(lst, reverse=True)
for i in (lst):
    x += 1
    print(features[x],"=", i)
```

```python
sns.set()
heats = sns.heatmap(X_train.corr())
```

# Обучаем и оцениваем

```python
dtrain = xgb.DMatrix(X_train, label=Y_train)
deval = xgb.DMatrix(X_eval, label=Y_eval)

param = {
    'max_depth': 10,  #10 the maximum depth of each tree
    'eta': 0.1,  # 0.3 the training step for each iteration
    'silent': 1,  # logging mode - quiet
#     'objective': 'binary:logistic',
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 2,
        }  # the number of classes that exist in this datset
num_round = 500 # the number of training iterations
```

```python
%%time
bst = xgb.train(param, dtrain, num_round)


from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
prob = bst.predict(deval)
pred_train = pd.DataFrame(np.asarray([np.argmax(line) for line in prob]))
print('Validation Set Accuracy:', accuracy_score(Y_eval, pred_train))

print('Validation Set ROC_AUC:', roc_auc_score(Y_eval, pred_train))
```

# Предсказываем

```python
preds = bst.predict(xgb.DMatrix(test[features][1:5221]))
# preds = bst.predict(xgb.DMatrix(X_train))
predictions = pd.DataFrame(np.asarray([np.argmax(line) for line in preds]))

predictions.hist()
print(predictions.shape)
from scipy.stats import itemfreq
itemfreq(predictions)
```

```python
submission=pd.DataFrame(predictions.values.ravel())

submission.describe()
```
